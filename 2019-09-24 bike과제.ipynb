```python
import sys
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp
import seaborn as sns
```

```python
# 데이터 준비
df = pd.read_csv('C:/Users/ktm/Anaconda_src/data/bike_train.csv')
```

```python
df.info()
```

```python
# 데이터 컬럼 나누기
# data에는 예측하고자 하는 'count'는 제외
# 'casual' , 'registered' 는 과적합 일어나므로 제외
# datetime 컬럼도 데이터 형태가 object라서 일단 제외
data = df.drop(['count','datetime', 'casual', 'registered'], axis=1)
```

```python
target = df['count']
```

```python
# sklearn 라이브러리에서 train_test_split 불러오기
from sklearn.model_selection import train_test_split
```

```python
# train_test_split으로 train 데이터, test 데이터를 나눈다.
X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=0)
```

```python
# sklearn에서 LinearRegression 을 불러오기
from sklearn.linear_model import LinearRegression
```

```python
# train 데이터로 LinearRegression 모델 만들기
model = LinearRegression().fit(X_train, y_train)
```

```python
# 만들어진 모델로 test 데이터 예측하기
pred = model.predict(X_test)
pred
```

```python
# 데이터 프레임 만들기(보기좋게)
dat = pd.DataFrame({'y_test':y_test, 'prediction':pred, 'diff':y_test-pred})
dat
```

```python
# MAE(mean absolute error 평균절대값 오차)
sum(abs(dat['diff']))/len(dat['diff'])
```

```python
# MSE(mean square error 평균 제곱 오차)
dat['diff']**2
```

```python
mse_value = sum(dat['diff']**2)/len(dat['diff'])
mse_value
```

```python
import numpy as np
```

```python
#RMSE ( root mean square error 평균 제곱근 오차)
result = np.sqrt(mse_value)
result
```

```python
# 훈련 데이터셋과 테스트 데이터셋 점수 확인
print('훈련 데이터 세트 점수: {:.2f}'.format(model.score(X_train, y_train)))
print('테스트 데이터 세트 점수: {:.2f}'.format(model.score(X_test, y_test)))
```
